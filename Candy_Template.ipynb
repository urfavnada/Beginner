# cleaning

### 4- drop all the useless columns and  rows with too much NaN ,
> tip : useless columns can be like `Internal ID`and can be also columns with too much NaN
- rename columns for easy access

```python
df.drop('Internal ID',axis=1,inplace=True)
df.drop('Unnamed: 113',axis=1,inplace=True)
```

### 5 - clean `going out` column
> tip : you should choose which best value to put in place of these NaNs ( mode, unknown,other ....etc)

```python
df["Q1: GOING OUT?"].isna().sum()
df["Q1: GOING OUT?"].fillna("unknown")
```

### 6 - clean ` gender` column
> tip : you should choose which best value to put in place of these NaNs ( mode, unknown,other ....etc)

```python
df["Q2: GENDER"].isna().sum()
df["Q2: GENDER"].unique()
df["Q2: GENDER"].fillna("Other")
df['Q2: GENDER'].replace("I'd rather not say",'Other',inplace=True)
```

### 7 - clean ` age` column
> tip : note the type of age column , and NaN values
 - try to use `to_numeric`
 - tip : you should choose which best value to put in place of these NaNs ( mean, median, ....etc)

```python
df["Q3: AGE"].unique()
df["Q3: AGE"].replace(['Old enough','Many','?','no','45-55','312','hahahahaha','older than dirt','5u','Enough',
                       'See question 2','24-50','Over 50','1000','ancient','OLD','old','1','MY NAME JEFF',
                       '59 on the day after Halloween','old enough','your mom','4',
                       'I can remember when Java was a cool new language','60+','46 Halloweens.',
                      ' 1000'],np.nan,inplace=True)
df["Q3: AGE"].replace("sixty-nine",69,inplace=True)
```

```python
pd.to_numeric(df['Q3: AGE'])
df['Q3: AGE'].describe()
```

```python
df['Q3: AGE'].fillna(42.6)
df = df[df['Q3: AGE'] < 100]
df["Q3: AGE"].unique()
```

### 8 - clean ` country` column
> tip : note that country column has country names with different writing ways
  - use fuzzywuzzy to deal with this problem
  - you should choose which best value to put in place of these NaNs ( mode, unknown,other ....etc)

```python
df["Q4: COUNTRY"].unique()
```

```python
df["Q4: COUNTRY"] = df["Q4: COUNTRY"].str.lower()
df["Q4: COUNTRY"] = df["Q4: COUNTRY"].str.strip()
df['Q4: COUNTRY'].replace(['canada','can','canae','canada`','soviet canuckistan'],"Canada", inplace = True)
df['Q4: COUNTRY'].unique()
```

```python
df['Q4: COUNTRY'].replace(['united states','us','united staes','united states of america',
                          'uae','u.s.a.','usausausa','america','us of a','unites states','the united states','unied states',
                          'u s','u s','the united states of america','unite states','usa? hard to tell anymore..',
                          'usas','united state','united sates','i pretend to be from canada, but i am really from the united states.',
                          'united stated','united statss','united ststes','usaa','usa! usa! usa!','n. america','ussa',
                          'u s a','united statea','usa usa usa!!!!','usa','murica','unhinged states',
                          'u.s.',"'merica",'new york','new jersey','north carolina','costa rica','california',
                          'mexico','cascadia','pittsburgh','murrika','ahem....amerca','alaska'],'USA',inplace = True)
df["Q4: COUNTRY"].unique()
```

```python
df['Q4: COUNTRY'].replace(['canada','can','canae','canada`','soviet canuckistan'],"Canada", inplace = True)
df['Q4: COUNTRY'].replace(['uk','united kingdom','england','u.k.','ireland','ud','endland'],"UK", inplace = True)
df['Q4: COUNTRY'].unique()
```

```python
df['Q4: COUNTRY'].replace(['finland','europe','netherlands','germany','greece','iceland','denmark','switzerland',
                          'the netherlands','spain','sweden','france'],'EU', inplace = True)
df['Q4: COUNTRY'].replace(['korea','japan','indonesia','singapore',
                           'china','taiwan','hong kong','south korea'],"Asia" , inplace = True)
df['Q4: COUNTRY'].unique()
```

```python
df['Q4: COUNTRY'].replace('australia','Australia', inplace = True)
df['Q4: COUNTRY'].replace('south africa','South africa', inplace = True)
df['Q4: COUNTRY'].replace('scotland','Scotland', inplace = True)
df['Q4: COUNTRY'].replace(['earth','trumpistan','narnia',"i don't know anymore",'insanity lately','a','atlantis','subscribe to dm4uz3 on youtube','fear and loathing'],np.nan, inplace = True)
df['Q4: COUNTRY'].value_counts()
```

### 10 - clean ` Q6` columns
>- tip : Q6 has all kinds of chocolate bars and questions with three distinct options to choose from (Joy, Meh, Despair).
- so all the column that has Q6 is a candy name , so you should go through all these column and you should choose which best value to put in place of these NaNs ( mode, unknown,other ....etc)

```python
# To deal with them all
Q6 = df.loc[:,[x for x in df.columns if 'Q6' in x]].copy()
Q6.fillna('MEH',inplace = True)
```

```python
# To know the most loved
Q6.replace({"JOY": 1, "MEH" : 0, "DESPAIR" : -1},inplace = True)
best = Q6.sum().sort_values(ascending = False).head(10)
best # The 10 most loved
```

```python
worst = Q6.sum().sort_values(ascending = True).head(10)
worst # the 10 most hated
```
