```python
# importing libiraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# read data
df = pd.read_csv("Mobile phone price.csv")
pd.set_option('display.max_rows', None)
df
```

```python
# To know the brands and its counts
# Q. What the brands count ?
df['Brand'].value_counts()
```

```python
# for nan values
# Q. Are there a nanvalue?
df.isna().sum()
```

```python
# To drop dupicated values
df.drop_duplicates(inplace=True)
```

```python
# Q. What the problem of price column?
# when i tried to convert price column to int i faced some problem
# 1. it was object type so i removed $
# 2. some rows has float types like this 1,299 so i replace , with ""
# then i convert the column type into int64
df["Price ($)"] = df["Price ($)"].str.replace(',', '', regex=True)
df["Price ($)"] = df["Price ($)"].str.strip("$")
df["Price ($)"] = df["Price ($)"].astype("int64")
```

```python
# Q. What the problem of storage and RAM columns
# i did the same thing to the Storage and Ram column
df["Storage "] = pd.to_numeric(df["Storage "].str.extract('(\d+)')[0])
df["RAM "] = pd.to_numeric(df["RAM "].str.replace("GB", ""))
# The name of the coulmns must be renamed
df.rename(columns={'Storage ': 'Storage'}, inplace=True)
df.rename(columns={'RAM ': 'RAM'}, inplace=True)
df.info()
```

```python
# To replace the spaces to _
df.columns = df.columns.str.replace(' ', '_')
df.info()
```

```python
# convert screen size column to int
# Q. What the problem in this coulmn
df["Screen_Size_(inches)"] = pd.to_numeric(df["Screen_Size_(inches)"], errors='coerce')
```

```python
# we had 2 rows that turned into nan value
df = df[df["Screen_Size_(inches)"] != "NaN"]
df["Screen_Size_(inches)"].count()
```

```python
# Clean Camera column
df["Camera_(MP)"] = df["Camera_(MP)"].str.replace("MP","")
df["Camera_(MP)"] = df["Camera_(MP)"].str.replace("D","")
df["Camera_(MP)"] = df["Camera_(MP)"].str.split("+")
listx = []
for i in df.index:
    listx.append(df["Camera_(MP)"][i][0])
df["Camera_(MP)"] = listx
df["Camera_(MP)"] = df["Camera_(MP)"].astype("float64")
```

```python
# To show the most used brands
plt.figure(figsize=(20, 10))
brand_counts = df['Brand'].value_counts().sort_values(ascending=False)
sns.barplot(x=brand_counts.index, y=brand_counts.values, color="purple")
plt.title('Brand Distribution Chart')
plt.xticks(rotation=90)
plt.xlabel('Brand')
plt.ylabel('Count')
plt.show()
```

```python
# To show the prices of the brands
plt.figure(figsize=(20, 10))
ordered_brands = df.groupby("Brand")["Price_($)"].median().sort_values(ascending=False).index
sns.barplot(x=df["Brand"], y=df["Price_($)"], color="purple", order=ordered_brands)
plt.title('Brand Distribution Chart')
plt.xticks(rotation=90)
plt.xlabel('Brand')
plt.ylabel('Count')
plt.show()
```

```python
# Create a plotly to know the battery capacity
fig = px.bar(
    df,
    x='Model',
    y='Battery_Capacity_(mAh)',
    title='Battery Capacity of Different Models',
    labels={'Model': 'Model', 'Battery_Capacity_(mAh)': 'Battery Capacity (mAh)'},
    color_discrete_sequence=['#800080'],
)

fig.update_layout(
    title_font=dict(size=20),
    xaxis=dict(tickfont=dict(size=10), tickangle=45),
    yaxis=dict(tickfont=dict(size=12)),
    plot_bgcolor='rgba(0,0,0,0)',  # Make the background transparent
)

# Order the x-axis by ascending battery capacity
fig.update_xaxes(categoryorder='total ascending')

fig.show()
```
